```python
def print_state(missionaries, cannibals, boat):
    print(f"Left Side: {missionaries} missionaries, {cannibals} cannibals")
    print(f"Right Side: {3 - missionaries} missionaries, {3 - cannibals} cannibals")
    print("Boat Position:", "Left" if boat == 0 else "Right")

def is_valid(missionaries, cannibals):
    if missionaries < 0 or missionaries > 3 or cannibals < 0 or cannibals > 3:
        return False
    if missionaries > 0 and missionaries < cannibals:
        return False
    if 3 - missionaries > 0 and 3 - missionaries < 3 - cannibals:
        return False
    return True

def check_win(missionaries, cannibals):
    return missionaries == 0 and cannibals == 0

def main():
    missionaries = 3
    cannibals = 3
    boat = 0  # 0 represents left, 1 represents right

    print("Welcome to Missionaries and Cannibals!")
    print("Rules:")
    print("- The boat can carry at most 2 people.")
    print("- Cannibals cannot outnumber missionaries on either side at any time.")
    print("\nLet's start!\n")

    while True:
        print_state(missionaries, cannibals, boat)

        if check_win(missionaries, cannibals):
            print("Congratulations! You've successfully solved the puzzle.")
            break

        print("\nEnter the number of missionaries and cannibals to move (separated by space), e.g., '2 1':")
        move = input().split()
        if len(move) != 2:
            print("Invalid input. Please provide two numbers.")
            continue

        try:
            move_missionaries = int(move[0])
            move_cannibals = int(move[1])
        except ValueError:
            print("Invalid input. Please provide valid numbers.")
            continue

        if move_missionaries + move_cannibals < 1 or move_missionaries + move_cannibals > 2:
            print("Invalid move. You can only move 1 or 2 people.")
            continue

        new_missionaries = missionaries - move_missionaries if boat == 0 else missionaries + move_missionaries
        new_cannibals = cannibals - move_cannibals if boat == 0 else cannibals + move_cannibals

        if not is_valid(new_missionaries, new_cannibals):
            print("Invalid move. The missionaries will be eaten.")
            continue

        missionaries = new_missionaries
        cannibals = new_cannibals
        boat = 1 - boat  # Toggle boat position

if __name__ == "__main__":
    main()
```
